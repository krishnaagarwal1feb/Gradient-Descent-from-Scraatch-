```
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  #define your equation here 
  return 0.005*x*x*x*x - 0.35*x*x - 0.4*x + 10
```

We could, in this simple case, compute the derivative, solve f′(x)=0, etc. But our goal is to understand gradient descent, so let’s do it!

There are 3 steps:

Take a random point x0.

Compute the value of the slope f′(x0).

Walk in the direction opposite to the slope: x1=x0−α∗f′(x0). Here, α is this learning rate we mentioned earlier. And the minus sign enables us to go in the opposite direction.

```
#here we take learning rate (alpha = 0.2) to start with 
#TASKS TO PERFORM  -- 
"""
        we will be finding the points where the minima would be located 
        we wil be finding the number of iterations that are used to achieve that point 
        Finally we will choose different learning rate [0.01(very small) , 10(very large)] and study the nature of the graph and plot the graphs 
"""
#run the algorithm until we get  : change in y < 0.001
```

```
#some points we will iterate on 
x = [0.2, 9.4, -1.1, -7.8]
```

```
def df(x):
  return 0.02*np.power(x,3) - 0.7*x - 0.4
```

```
slope = df(x[0]);slope
```

```
lr = 0.2  #learning rate 
```

```
for i in range(4):
  x.append(x[i] - lr*df(x[i])) 
```

```
y = []
x = []
for x_iter in np.arange(-8,9.5,0.1):
  x.append(x_iter)
  y.append(f(x_iter))
plt.plot(x,y)
#this is the equation plotted in the given range
```

```
plt.plot(x,y)
x1 = [0.2, 9.4, -1.1, -7.8]   #set of initial values
for i in x1:
  plt.plot(i,f(i), 'ro')
plt.title("Initial points over the quations")
plt.xlabel('X')
plt.ylabel('y')
```

```
plt.plot(x,y)
x0 = x1[0]
y0 = f(x0) 
iterations = 0
plt.plot(x0,y0,'ro')
while True:
  #here we need to find the new x and y 
  x_new = x0 - lr*df(x0)
  y_new = f(x_new)
  plt.plot(x_new, y_new, 'ro')
  iterations += 1
  if abs(y_new - y0) <= 0.001:
    break
  x0 = x_new
  y0 = y_new
plt.xlabel('X')
plt.ylabel('y')
plt.title('Plotting the first iteration with lr=0.2 with iterations=' + str(iterations))
```

```
plt.plot(x,y)
x0 = x1[1]
y0 = f(x0) 
iterations = 0
plt.plot(x0,y0,'go')
while True:
  #here we need to find the new x and y 
  x_new = x0 - lr*df(x0)
  y_new = f(x_new)
  plt.plot(x_new, y_new, 'go')
  iterations += 1
  if abs(y_new - y0) <= 0.001:
    break
  x0 = x_new
  y0 = y_new
plt.xlabel('X')
plt.ylabel('y')
plt.title('Plotting the first iteration with lr=0.2 with iterations=' + str(iterations))
```

```
print("Value of first minima is : ", (x_new, y_new))
```
